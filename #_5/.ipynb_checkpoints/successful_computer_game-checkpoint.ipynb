Следующий столбец - `critic_score` (оценка критиков, максимум 100 баллов). 
Данные в колонке представлены в виде числа с плавающей точкой `float64`. По логике, они должны иметь целочисленный вид.<br>Однако, по причине наличия в колонке значений `NaN` (англ. Not-a-Number, «не число» — одно из особых состояний числа с плавающей запятой), придать этой колонке требуемый тип `int64` не представляется возможным. Машина выдает ошибку: `IntCastingNaNError: Cannot convert non-finite values (NA or inf) to integer`.<br>А пропусков в этой колонке очень много - значит, удалять строки с пропусками в этой колонке нецелесообразно:

```python
df['critic_score'].isna().sum()
```

Приведем значения в колонке к целочисленному виду. Проверим.

```python
df['critic_score'] = df['critic_score'].astype('int64')
df['critic_score'].dtypes
#df.info()
```

Оценим уникальные значения. 

```python
sorted(df['critic_score'].unique())
```

### Предобработка данных в колонке ` user_score `         

Следующий столбец - `user_score` (оценка пользователей, максимум 10 баллов). 
Как мы видим из общей информации, данные в колонке представлены в виде `object` (текстовые или, как в нашем случае, - смешанные числовые и нечисловые значения).<br>В столбце много пропусков.

```python
df['user_score'].isna().sum()
```

Кроме того, обратим внимание на такое значение в колонке `user_score`, как `tbd`.

```python
df['user_score'].unique()
```

```python
df[df['user_score'] == 'tbd'].head(10)
```

Таких значений `tbd` в нашем столбце очень много.

```python
# len(df.loc[df['user_score'] == 'tbd'])
tbd = 'tbd'
len(df.query('user_score in @tbd'))
```

`tbd`  в переводе с англ. "to be determined" означает "будет определено" (в смысле "будет определено позже").<br>Для нас на момент исследования датасета слова "будет определено позже" не имеют смысла. Сейчас игры с оценкой `tbd` просто не имеют рейтинга. Заменим `tbd` на `NaN`.<br>И далее, как и в предыдущем случае, заменим все пропуски значением `0`. Будем знать, что в этом случае рейтинг отсутствует.

```python
# df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'] = df['user_score'].fillna(0)
df['user_score'].isna().sum()
```

```python
df['user_score'].unique()
```

В соответствии с описанием проекта, данные должны иметь вид чисел с плавающей точкой в интервале от 0 до 10.<br>Используем метод `pd.to_numeric()`, чтобы перевести значения столбца из строкового типа в тип `float64`.<br>Думается, что целесообразно будет привести значения в этом столбце и в столбце `critic_score` к одной шкале и  перевести значения столбца `user_score` из типа `float64` в тип `int64`. 

```python
df['user_score'] = pd.to_numeric(df['user_score'], errors ='coerce')
df['user_score'] = df['user_score'] *10
df['user_score'] = df['user_score'].astype('int64')
sorted(df['user_score'].unique())
```

```python
df['user_score'].dtypes
# df.info()
```

Таким образом, все значения в этом столбце находятся в диапазоне от 0 до 100, что соответствует логике проекта.<br>Перейдем к следующей колонке.

### Предобработка данных в колонке ` rating `         

Следующий столбец - `rating` (рейтинг от организации ESRB (англ. Entertainment Software Rating
Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию). 
Как мы видим из общей информации, данные в колонке представлены в виде `object`.<br>В столбце много пропусков.

```python
df['rating'].isna().sum()
```

```python
df['rating'].unique()
```

Чтобы подступиться к предварительной обработке значений в столбеце `rating`, сначала нам нужно понять, какой смысл скрывается за этими аббревиатурами. Обратимся к Википедии :) 
- «E» («Everyone») — «Для всех»;
- «M» («Mature») — «Для взрослых»;
- «T» («Teen») — «Подросткам»;
- «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»;
- «K-A» («Kids to Adults») - «От детей к взрослым». Так первоначально назывался рейтинг «E» («Everyone») — «Для всех»;
- «AO» («Adults Only 18+») — «Только для взрослых»;
- «EC» («Early childhood») — «Для детей младшего возраста»; 
- «RP» («Rating Pending») — «Рейтинг ожидается».

Теперь нам понятно, что рейтинг «K-A» нужно переименовать в рейтинг «E».<br>Как и в предыдущем пункте проекта, для нас на момент исследования датасета рейтинг «RP» — «Рейтинг ожидается» не имеет смысла. Заменим `RP` на `NaN`. 

```python
df['rating'] = df['rating'].replace('K-A', 'E')
df.loc[df['rating'] == 'RP', 'rating'] = np.nan
# df['rating'] = df['rating'].replace('RP', np.nan)
```

Теперь заменим все пропуски `NaN` на `unknown`.

```python
df['rating'] = df['rating'].fillna('unknown')
```

Приведем значения в колонке к нижнему регистру.

```python
df['rating'] = df['rating'].str.lower()
df['rating'].unique()
```

### Анализ периода (срока) жизни платформ         

Учебный проект предлагает нам ответить на вопрос: за какой характерный срок появляются новые и исчезают старые платформы? То есть, предлагается определить период (срок) жизни платформ на рынке компьютерных игр.

Проведем анализ по всему первоначальному датасету `df`.

```python
# df.head()
```

```python
lifetime_platform = df.pivot_table(index=['platform'], values='year_of_release', aggfunc=('min', 'max'), fill_value=0)
lifetime_platform = pd.DataFrame(lifetime_platform)
lifetime_platform.reset_index(drop = False, inplace = True)
lifetime_platform['life_cycle'] = lifetime_platform['max'] - lifetime_platform['min']
lifetime_platform.sort_values('life_cycle', ascending=False)
```

Наблюдаем, что есть игровые платформы, которые нисколько не задержались на рынке, и есть платформа `pc`. Тридцать один год на рынке!

```python
sns.set(rc={'figure.figsize':(20, 10)})
sns.barplot(data=lifetime_platform, x='platform', y='life_cycle')
plt.title('Период (срок) жизни платформ на рынке компьютерных игр', fontsize=16)
plt.xlabel('игровые платформы')
plt.ylabel('период (срок) жизни');
```

А в среднем жизненный цикл (точнее, срок присутствия на рынке) игровой платформы составляет...

```python
# lifetime_platform['life_cycle'].median()
lifetime_platform['life_cycle'].mean()
```

...7 лет и 7 месяцев.

## Проверка гипотез

Учебное задание предлагает проверить следующие гипотезы:
- средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- средние пользовательские рейтинги жанров Action и Sports разные.

В этом задании мы будем использовать первоначальный датафрейм `df`. Он будет являться генеральной совокупностью.

```python
# df.head()
```

### Проверка гипотезы: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

Начнем с УЧЕБНОЙ гипотезы о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем Нулевую гипотезу:  в генеральной совокупности `df` средний пользовательский рейтинг платформы `xone` равен среднему пользовательскому рейтингу платформы `pc`.

Cформулируем Альтернативную гипотезу:  в генеральной совокупности `df` средний пользовательский рейтинг платформы `xone` НЕ равен среднему пользовательскому рейтингу платформы `pc`.

Далее на основе полученного значения `p-value` мы сделаем выбор - отклонить / не отклонять Нулевую гипотезу в пользу Альтернативной гипотезы.

Сформируем два датасета `xone_in_df` и `pc_in_df`, которые будут являться нашими стратифицированными выборками по отношению к генеральной совокупности `df`.

```python
xone = 'xone'
xone_in_df = df.query('platform == @xone')
# xone_in_df = df.loc[df['platform'] == 'xone']
xone_in_df.reset_index(drop = True, inplace = True)
xone_in_df.head()
```

```python
pc = 'pc'
pc_in_df = df.query('platform == @pc')
# pc_in_df = df.loc[df['platform'] == 'pc']
pc_in_df.reset_index(drop = True, inplace = True)
pc_in_df.head()
```

Зададим критическое значение уровня значимости проверки гипотезы 5%. Наша Нулевая гипотеза является двусторонней. Нулевая гипотеза будет отвергнута, если вероятность получить выборочное среднее будет меньше 5%.

```python
alpha = 0.05
```

Поместим обе выборки в метод `scipy.stats.ttest_ind (array1, array2, equal_var)`.

```python
# alpha = 0.05
results = st.ttest_ind(xone_in_df['user_score'], pc_in_df['user_score'], equal_var = False )
results.pvalue
```

Формально:

```python
if results.pvalue < alpha:
    print('Отвергаем Нулевую гипотезу')
else:
    print('НЕ Отвергаем Нулевую гипотезу')
```

Формально мы отвергаем Нулевую гипотезу в пользу сформулированной Альтернативной гипотезы: в генеральной совокупности `df` средний пользовательский рейтинг платформы `xone` НЕ равен среднему пользовательскому рейтингу платформы `pc`.

Проверим:

```python
# xone_in_df['user_score'].mean()
xone_in_df['user_score'].describe()
```

```python
# pc_in_df['user_score'].mean()
pc_in_df['user_score'].describe()
```

<b>Вывод</b>. УЧЕБНУЮ гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, отвергаем.
